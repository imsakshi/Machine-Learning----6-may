# Assumptions of Linear Regression

```python
1. Linear Distribution : It is defined as a relationship between two features where change in one feature
    can easily explain change in another feature.
    There should be a linear relationship in between Target Variable and Independent Features.
    To check : we simply draw a scatter plot.
    
```

```python
2. Presence of Normality: It simply means that all the features that will be a part of "X"
    Feature Marix should obey a normal distribution.
    To Check : We make a use of Histogram.
```

```python
3. MultiCollinearity : is defined as the correlation between the features used for regresssion analysis.
    Correlation is a statistical parameter used to find the relationship between target 
    variable and all independent features.
    Relationships:
        1. positive relationship
        2. negative relationship
        3.no relationship.
It is a measure of correlation among all the columns present in "X" matrix.
For a regression analysis,we dont want the features to be heavily dependent on one another as the 
changing one might change the another.
We need very little or no Multicollinearity.
To check : Pearsons correlation coefficient or Heat-map.
```

```python
4. AutoCorrelation : It can be defined as correlation between adjacent observations in the vector
                     of prediction(Dependent variable).
Sometimes the value of y(x+1) is dependent on the value of y(x) which again depends upon the value of y
y(x-1).
Common Type of datasets : Time-series Analysis problem or stock Market data.
To check For AC: line plots or gemetrical plots.
```

```python
5.HomoScedasticity : (Homo : similar | scedasticity : error/noise)
It can be defined as a property of regression models where errors/noise or random distribution between 
target variable and all independent features is similar.
If erors keep changing drastically, this will result in a funnel shaped scatter plots and break our 
regression model condition.
To check it: Scatter plots
```
