```python
Classification Problem : KNN(k-nearest neighbor)
It comes under the category of lazy algorithms.
```

```python
Knn algorithm is a method for classifying objects based on closest training examples in the feature space.
Instance based learning
It is the simplest and fundamental technique where there is no prior knowledge about the distribution of data 
is required.
```

```python
Algorithm : for all unknown samples
                 for all known samples
                     compute distance between sample and new-instance
                 end for
                find the k- smallest distances
                assign new-instance(i) to the class which appears more frequently.
            end for.
```

```python
The performance of knn depends on:
    1. The value of k.
    2. Distance metric applied.
```

```python
There is no hard n fast rule to choose the value of k but still one of the mostky
used technique to find the value of k , 
  k = underoot of n
    n is your no of observations.
    if i have 100 observations k = 10
    It is also preferable to take the value of k is an odd number.
2. Euclidean distance or manhattan distance.
```

```python
small value of k means that noise will have higher influence on the result.
Higher value of k make its computationally expensive.
```

```python
1. standardization : it is a scaling technique where the values are centered around the mean with a unit standard deviation
 This means that the mean of the attribute becomes zero and resultant distribution has a unit standard 
deviation.
```

```python
Min-max Scaler : It transforms the features by scaling each feature to a given range. the range can be specified by 
    feature_range parameter. default at (0,1). 
Note: It is very sensitive to outliers, so if there are outliers in the data, you might want to consider another 
    scaler.
    
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
Robust scaler :
     if your data contains many outliers, scaling using mean and std dev
        of the data is likely to not work well. 
In such cases, we can use robust scaler.
It removes the median and scales the data according to
quantile range.
By default. the scaler used the inter quartile range(IQR) , which is the
range between 1st quartile and 3rd quartile.
```

```python
from sklearn.preprocessing import RobustScaler
robust = RobustScaler()
```
